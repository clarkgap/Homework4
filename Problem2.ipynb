$
\begin{aligned}
\min_{x_1,x_2} & \quad  f=-x_1\\
{\text{subject to }} & \quad g_1=x_2-(1-x_1)^3\leq 0{\quad} {\rm and}{\quad} -x_2\leq 0.
\end{aligned}
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Contour Line
delta = 0.001
x1 = np.arange(-0.0, 1.5, delta)
x2 = np.arange(-0.5, 1.0, delta)
X1, X2 = np.meshgrid(x1, x2)
Y = -X1

cp = plt.contour(X1, X2, Y)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('X1')
plt.ylabel('X2')

# Feasible domains
plt.imshow( ((X2-(1-X1)**3<=0) & (-X2<=0)).astype(int) ,
                extent=(X1.min(),X1.max(),X2.min(),X2.max()),origin="lower", cmap="Greys", alpha = 0.3)
# Boundaries
plt.plot(x1, (1-x1)**3, label='$g_1=x_2-(1-x_1)^3\leq 0$')
plt.plot(x1, 0*np.ones_like(x1), label='$g_2=-x_2\leq 0$')
plt.legend(bbox_to_anchor=(1.0, 1), loc='upper left')

# Solution
plt.plot(1.0,0.0, marker="o", color="black",markersize=10)

plt.show()
```
