$
\begin{aligned}
\min_{x_1,x_2} & \quad f({\bf x})=(x_1+1)^2+(x_2-2)^2\\
{\text{subject to }} & \quad g_1 = x_1-2\leq 0,{\quad} g_3 = -x_1\leq 0,\\
& \quad g_2 = x_2-1\leq 0, {\quad} g_4 = -x_2\leq 0.
\end{aligned}
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Contour Line
delta = 0.01
x1 = np.arange(-1.0, 3.0, delta)
x2 = np.arange(-1.0, 2.0, delta)
X1, X2 = np.meshgrid(x1, x2)
Y = (X1+1)**2 + (X2-2)**2

cp = plt.contour(X1, X2, Y)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('X1')
plt.ylabel('X2')

# Feasible domains
plt.imshow( ((X1-2<=0) & (X2-1<=0) & (-X1<=0) & (-X2<=0)).astype(int) ,
                extent=(X1.min(),X1.max(),X2.min(),X2.max()),origin="lower", cmap="Greys", alpha = 0.3)
# Boundaries
plt.axvline(x=2, color='b', label='g1 = X1-2 $\leq$ 0')
plt.plot(x1, 1*np.ones_like(x1), label='g2 = X2-1 $\leq$ 0')
plt.axvline(x=0, color='r', label='g3 = -X1 $\leq$ 0')
plt.plot(x1, 0*np.ones_like(x1), label='g4 = -X2 $\leq$ 0')
plt.legend(bbox_to_anchor=(1.0, 1), loc='upper left')

# Solution
plt.plot(0.0,1.0, marker="o", color="black",markersize=10)

plt.show()
```
